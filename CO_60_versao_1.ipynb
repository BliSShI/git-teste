```python
# Inicializa uma lista vazia para armazenar os valores da primeira coluna
primeira_coluna = []

# Abre o arquivo de texto para leitura
with open('dados.txt', 'r') as file:
    # Lê todas as linhas do arquivo
    linhas = file.readlines()
    # Itera sobre as linhas do arquivo
    for linha in linhas:
        # Divide a linha em colunas usando o espaço como delimitador
        colunas = linha.split()
        # Extrai o valor da primeira coluna e adiciona à lista
        primeira_coluna.append(float(colunas[8]))

# Imprime os valores da primeira coluna
print(primeira_coluna)
```

```python

### Cálculo da dose
n = len(primeira_coluna)
dose = []
for i in range(0, n):
  dose.append(primeira_coluna[i]*fluence)

print('Dose máxima =', max(dose))
print(n)
print(profundidade[0])

pdp = []

for i in range(0, n):
  pdp.append(dose[i]*100/max(dose))

profundidade.pop(0)
pdp.pop(0)
plt.scatter(profundidade, pdp, label='SSD = 80 cm \nCubo convencional')

plt.title('Porcentagem de Dose em Profundidade')

plt.legend()

plt.xlabel('Profundidade (cm)')
plt.ylabel('%')

plt.ylim(-5, 105)

plt.show()
```

```python
aux=0
for i in profundidade:
  print('pdp(%.2f) =' %i, '%.2f' %pdp[aux])
  aux+=1
```

```python
def funcsg(xc):
    ''' Definimos as funcoes g que serao usadas '''
    n = len(xc)

    g1c = np.array([1.0]*n)
    g2c = np.array(xc)
    g3c = np.array(xc)**2

    gm = [g1c, g2c, g3c]

    return gm

def matrizAeb(xc,fc,dim):
    m = len(xc)

    G = funcsg(xc)  ### Obtem as funcoes base nos pontos x_k

    bc = []
    Ac = np.empty((dim,dim))  ### matriz A inicial vazia

    #### Calculo matriz A

    for i in range(1,dim+1,1):
      for j in range(1,dim+1,1):
        soma = 0.0
        for k in range(0,m,1):
          soma = soma + (G[i-1][k]*G[j-1][k])

        Ac[i-1][j-1] = soma


    #### Vetor b
    for i in range(1,dim+1,1):
      soma = 0.0
      for k in range(0,m,1):

        soma = soma + (fc[k]*(G[i-1][k]))

      bc.append(soma)

    return Ac, bc
```

```python
#### Dados tabelados
x = profundidade
f = pdp

A, b = matrizAeb(x,f,3)   #### Obtencao da matriz A e vetor b

print('b = ', b)
print(' A = ', A)

#### Solucao do sistema linear A*alpha = b usando 'numpy'
coeff = np.linalg.solve(A, b)  ### alphas

print('Coeficientes = ', coeff)

#### Grafico
xm = np.linspace(x[0],x[-1],200)

phi = []

for i in range(len(xm)):
  phi.append(coeff[0] + coeff[1]*xm[i] + coeff[2]*xm[i]**2)

plt.scatter(x,f, color='blue')
plt.plot(xm,phi,color='red')
plt.grid()
plt.show()
```
